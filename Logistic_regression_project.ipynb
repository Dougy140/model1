```python
import pandas as pd
import numpy as np
```

```python
pima = pd.read_csv('pima.csv')
pima.head()
```

```python
column = ['pregnant','glucose','bp','skin','insulin_level','bmi','pedigree','age','diabetes_label']
```

```python
pima = pd.read_csv('pima.csv',names=column)
pima.head()
```

```python
pima.tail()
```

```python
feature = ['pregnant','insulin_level','bmi','age']
```

```python
X = pima[feature]
y = pima['diabetes_label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
prediction = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,prediction))
```

```python
y_test.value_counts()
```

```python
y_test.mean()
```

```python
print("     True diabetes label : ", y_test.values[0:15])
print("Predicted diabetes label : ", prediction[0:15])
```

```python
#confusion matrix
print(metrics.confusion_matrix(y_test,prediction))
```

```python
#error = (FP+FN)/(TP+TN+FP+FN )
```

```python
#Sensitivity
print(metrics.recall_score(y_test,prediction))
```

```python
#precision
print(metrics.precision_score(y_test,prediction))
```
